```python
# Import the necessary packages and give them aliases
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a dataframe variable called df by calling the read_csv() function of the pandas package
myDF = pd.read_csv('IA breast cancer v2d.csv')
```

```python
myDF.head()
```

```python
myDF['diagnosis'].value_counts()
```

```python
myDF.drop('id', axis=1, inplace=True)
myDF
```

```python
# Encoding the diagnosis column
myDF['diagnosis'] = myDF['diagnosis'].map({'M': 1, 'B': 0})
missing_data = myDF.isnull().sum()
missing_data
```

```python
from sklearn.impute import SimpleImputer

# 2.3 Handle missing data
imputer = SimpleImputer(strategy='mean')
myDF['concavity_se'] = imputer.fit_transform(myDF[['concavity_se']])

# Verify if the missing value has been imputed
missing_data_after_imputation = myDF.isnull().sum()
missing_data_after_imputation
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Define features and target variable
X = myDF.drop('diagnosis', axis=1)
y = myDF['diagnosis']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize scalers
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()

# Scale data using StandardScaler
X_train_standard = scaler_standard.fit_transform(X_train)
X_test_standard = scaler_standard.transform(X_test)

# Scale data using MinMaxScaler
X_train_minmax = scaler_minmax.fit_transform(X_train)
X_test_minmax = scaler_minmax.transform(X_test)

# Checking the shape of the scaled data to ensure correctness
(X_train_standard.shape, X_test_standard.shape, X_train_minmax.shape, X_test_minmax.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Initialize Logistic Regression model
logreg = LogisticRegression()

# Train models on the two sets of data
# Using StandardScaler data
logreg_standard = logreg.fit(X_train_standard, y_train)
y_pred_standard = logreg_standard.predict(X_test_standard)

# Using MinMaxScaler data
logreg_minmax = logreg.fit(X_train_minmax, y_train)
y_pred_minmax = logreg_minmax.predict(X_test_minmax)

```

```python
# calculate accuracy, precision, recall, f1-score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Accuracy
accuracy_standard = accuracy_score(y_test, y_pred_standard)
accuracy_minmax = accuracy_score(y_test, y_pred_minmax)

# Precision
precision_standard = precision_score(y_test, y_pred_standard)
precision_minmax = precision_score(y_test, y_pred_minmax)

# Recall
recall_standard = recall_score(y_test, y_pred_standard)
recall_minmax = recall_score(y_test, y_pred_minmax)

# F1-score
f1_standard = f1_score(y_test, y_pred_standard)
f1_minmax = f1_score(y_test, y_pred_minmax)

# Group the scores into a dataframe
scores = pd.DataFrame({'Standard Scaler': [accuracy_standard, precision_standard, recall_standard, f1_standard],
                       'MinMax Scaler': [accuracy_minmax, precision_minmax, recall_minmax, f1_minmax]},
                      index=['Accuracy', 'Precision', 'Recall', 'F1 Score'])
scores
```
